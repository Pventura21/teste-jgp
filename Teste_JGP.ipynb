```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
link = 'https://search.brave.com/search?q=cota%C3%A7%C3%A3o+dolar'
requisicao = requests.get(link)
```

```python
print(requisicao)
```

```python
site = BeautifulSoup(requisicao.text, 'html.parser')
```

```python
print(site.prettify())
```

```python
pesquisa = site.find('input', class_='desktop-large-regular svelte-hjp3ip')
print(pesquisa['value'])
```
